```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
```

```python
dataset = pd.read_csv("loan_prediction.csv")
```

```python
print(dataset.head())
```

```python
#define X
X = dataset[["LoanAmount","ApplicantIncome"]]

#Visualise data points
plt.scatter(X["ApplicantIncome"],X["LoanAmount"])
plt.xlabel('AnnualIncome')
plt.ylabel('Loan Amount (In Thousands)')
plt.show()
```

```python
#Number of Clusters
kmeans = KMeans(n_clusters=3)

#Fitting the input data
kmeans.fit(X)

#Getting the cluster lables
lables = kmeans.predict(X)

#Centroid Values
centroids = np.array(kmeans.cluster_centers_)

print("Centroid values")
print(centroids)
```

```python
#no. of Clusters 
K = 3

#plotting the clusters
plt.scatter(X["ApplicantIncome"],X["LoanAmount"],c=kmeans.labels_, cmap='rainbow')
plt.xlabel('AnnualIncome')
plt.ylabel('Loan Amount (In Thousands)')
plt.show()
```

```python
#plotting centroids
plt.scatter(X["ApplicantIncome"],X["LoanAmount"],c=kmeans.labels_, cmap='rainbow')
plt.scatter(centroids[:,1], centroids[:,0], marker="x", color='black')
plt.xlabel('AnnualIncome')
plt.ylabel('Loan Amount (In Thousands)')
plt.show()
```
