```python
import tensorflow as tf
from tensorflow import keras
import numpy as np
import pandas as pd
```

```python
observations= {}
```

```python
#creating array X and Y such that Y = (X/2) - 4
X = np.array([])
Y = np.array([])
for i in range (0, 11, 2):
    X = np.append(X,i)
    y = (i / 2) - 4
    Y = np.append(Y,y)
    
print(X)
print(Y)
```

```python
#defining a single layer; single neuron model
model = tf.keras.Sequential([keras.layers.Dense(units=5, input_shape=[1]), keras.layers.Dense(units=1)])

#defining the model evaluation for backpropogation
model.compile(optimizer='sgd', loss='mean_squared_error')

model.fit(X, Y, epochs=300)
```

```python
observations["PERCEPTRON: [1-5-1-1] OPTIMIZER: 'sgd' LOSS: 'mean_squared_error' EPOCH: 300"]=(model.predict([40.]))
```

```python
for key, value in observations.items():
    print(key, " ===> ", value)
```

```python
#defining a single layer; single neuron model
model = tf.keras.Sequential([keras.layers.Dense(units=3, input_shape=[1]), keras.layers.Dense(units=2, input_shape=[1]), keras.layers.Dense(units=1, input_shape=[1]) ])

#defining the model evaluation for backpropogation
model.compile(optimizer='rmsprop', loss='mean_squared_logarithmic_error')

model.fit(X, Y, epochs=2500)
```

```python
observations["PERCEPTRON: [1-3-2-1-1] OPTIMIZER: 'rmsprop' LOSS: 'mean_squared_logarithmic_error' EPOCH: 2500"]=(model.predict([40.]))
```

```python
for key, value in observations.items():
    print(key, " ===> ", value)
```
